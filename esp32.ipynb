```python
from langchain_community.document_loaders import PyPDFLoader

# Load your PDF
loader = PyPDFLoader("esp32_datasheet_en.pdf")
txtData = loader.load()
print(txtData[0])
```

```python
# Splitting data into chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,      # Each chunk will be ~200 characters
    chunk_overlap=20     # Overlap of 20 characters between chunks
)

# Split the loaded PDF documents into smaller chunks
final_documents = text_splitter.split_documents(txtData)

# Inspect the first few chunks
for i, doc in enumerate(final_documents[:5]):
    print(f"Chunk {i+1}:\n{doc.page_content}\n")
```

```python
# data embedding means converting txt data into vectors
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
# vector storage
from langchain_community.vectorstores import FAISS

db = FAISS.from_documents(
    final_documents,   # <-- use the correct variable name
    embeddings
)

db


```

```python
# retrieval (similarity search)
query = input("Enter your question for cpp: ")
docs = db.similarity_search(query)

for i in range(len(docs)):
    print(docs[i].page_content)
```
